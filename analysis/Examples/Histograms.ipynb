```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
df.head()
```

```python
sns.distplot( df["sepal_length"] ) # around 8 bins 
sns.plt.show()
```

```python
sns.distplot( df["sepal_length"], bins=20)
sns.plt.show()
```

```python
sns.distplot( a=df["sepal_length"],hist=True, kde=False) # no density curve
# kde = kernel density estimation
# estimates the probability density function
sns.plt.show()
```

```python
sns.distplot( a=df["sepal_length"], hist=True, kde=True, rug=True )
# rug plots illustrate marginal distribution of a variable
sns.plt.show()
```

```python
rug_args = {
    "color": "r",
    "alpha": 0.3,
    "linewidth": 2,
    "height":0.2
}
sns.distplot( a=df['sepal_length'], rug=True, rug_kws=rug_args)
sns.plt.show()
```

```python
kde_args = {
    "color": "g",
    "alpha": 0.3,
    "linewidth": 5,
    "shade": True
}
sns.distplot( a=df['sepal_length'], kde=True, kde_kws=kde_args)
sns.plt.show()
```

```python
sns.distplot( a=df['sepal_length'], kde=True, \
             kde_kws=kde_args, vertical=True)
sns.plt.show()
```

# Boxplot with Histogram

```python
f, (ax_box, ax_hist) = plt.subplots(2, \
        sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
# unpacks the axes array immediately
# axes array is 1 dimensional
```

```python
f, ax_box, ax_hist
```

```python
sns.boxplot(df["sepal_length"], ax=ax_box)
sns.distplot(df["sepal_length"], ax=ax_hist)
```

```python
# remove the x axis label for the top and try again
ax_box.set(xlabel='')
```

# Histogram with Several Variables

```python
sns.distplot( df["sepal_length"] , color="skyblue", label="Sepal Length")
sns.distplot( df["sepal_width"] , color="red", label="Sepal Width")
sns.plt.legend() # Shows a legend, with appropriate label and color
sns.plt.show()
```

```python
f, axes = plt.subplots(2, 2, \
        figsize=(7, 7),
        sharex=True) #2x2 plot, plots share x axis markings
sns.distplot( df["sepal_length"] , color="skyblue", ax=axes[0, 0])
sns.distplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( df["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( df["petal_width"] , color="teal", ax=axes[1, 1])
sns.plt.show()
```

```python
f, axes = plt.subplots(2, 2, \
        figsize=(10, 7), # increased the width, but height constant
        sharex=True) #2x2 plot, plots share x axis markings
sns.distplot( df["sepal_length"] , color="skyblue", ax=axes[0, 0])
sns.distplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( df["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( df["petal_width"] , color="teal", ax=axes[1, 1])
sns.plt.show()
```

```python
f, axes = plt.subplots(2, 2, \
        figsize=(10, 7), # increased the width, but height constant
        sharex=False) #2x2 plot, plots do not share x axis markings
sns.distplot( df["sepal_length"] , color="skyblue", ax=axes[0, 0])
sns.distplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( df["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( df["petal_width"] , color="teal", ax=axes[1, 1])
sns.plt.show()
```
