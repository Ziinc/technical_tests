# One numerical variable

```python
import seaborn as sns

df = sns.load_dataset('iris')
df.head()
```

```python
df.species.unique()
```

```python
df.count()
```

```python
df.dtypes
```

```python
sns.boxplot( y=df['sepal_length'])
sns.plt.show()
```

## Several Groups

```python
sns.boxplot( x=df['species'], y=df['sepal_length'])
sns.plt.show()
```

# Several numerical variables

```python
# show everything on x axis, and only index 0 to 2 (noninclusive) on y axis
sns.boxplot(data=df.iloc[:,0:2]) 
sns.plt.show()
```

```python
# show everything on x axis, and only index 0 to 3 (noninclusive) on y axis
sns.boxplot(data=df.iloc[:,0:3]) 
sns.plt.show()
```

# Horizontal Boxplot

```python
sns.boxplot( y=df['species'], x=df['sepal_length'])
sns.plt.show()
```

# Customizing the boxplot

```python
sns.boxplot( x=df['species'], y=df['sepal_length'], linewidth=5)
sns.plt.show()
```

```python
sns.boxplot( x=df['species'], y=df['sepal_length'], 
            linewidth=5,
           notch=True)
sns.plt.show()
```

```python
sns.boxplot( x=df['species'], y=df['sepal_length'], 
            linewidth=5,
           notch=True,
           width=0.5)
sns.plt.show()
```

## Colours

```python
sns.boxplot( x=df['species'], y=df['sepal_length'], 
            linewidth=5,
           notch=True,
           width=0.5,
           palette="Blues")
sns.plt.show()
```

## Adding notches

```python
sns.boxplot( x=df['species'], y=df['sepal_length'], 
            linewidth=5,
           notch=True,
           width=0.5,
           color="skyblue")
sns.plt.show()
```

## Adding custom color to each boxplot

```python
palette = {
    "versicolor": "g", #green
    "setosa": "b", #blue
    "virginica": "m", #magenta
}
sns.boxplot( x=df['species'], y=df['sepal_length'], 
            linewidth=5,
           notch=True,
           width=0.5,
           palette=palette)
sns.plt.show()
```

## conditional highlighting

```python
palette = {
    species: "r" if species == "versicolor" else "b" \
        for species in df.species.unique()
}
sns.boxplot( x=df['species'], y=df['sepal_length'], 
            linewidth=5,
           notch=True,
           width=0.5,
           palette=palette)
sns.plt.show()
```

## Adding transparency

```python
palette = {
    species: "r" if species == "versicolor" else "b" \
        for species in df.species.unique()
}

ax = sns.boxplot( x='species', y='sepal_length', data=df)

for patch in ax.artists:
    r, g, b, a = patch.get_facecolor()
    patch.set_facecolor((r,g,b,0.3))
sns.plt.show()
```

# Grouped Boxplot

```python
df_tips = sns.load_dataset('tips')
df_tips.head()
```

```python
sns.boxplot(x='day', y='total_bill', hue="smoker",\
            data = df_tips, palette="Set1")
sns.plt.show()
```

## Control Order of boxplot

```python
sns.boxplot(x='species', y='sepal_length',\
            data=df, order=["virginica","setosa", "versicolor"])
sns.plt.show()
```

```python
order = df.groupby(by=['species'])['sepal_length'].median()
order
```

```python
order = order.iloc[::-1].index #indexes by decreasing mean
order
```

```python
sns.boxplot(x='species', y='sepal_length', data=df, order=order)
sns.plt.show()
```

## Add  jitter

```python
ax = sns.boxplot(x='species', y='sepal_length', data=df)
ax = sns.swarmplot(x='species', y='sepal_length', data=df, color="grey")
sns.plt.show()
```
