# basic sine graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
      title = 'Simple sine plot')
ax.grid()
plt.show()
```
